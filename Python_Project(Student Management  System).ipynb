```python
from tkinter import *
import mysql.connector
from tkinter import messagebox
from PIL import ImageTk,Image

########## Title#######

window = Tk()
window.geometry("500x650")
window.minsize(500,650)
window.maxsize(500,650)
window.title('Welcome to the Student Management System ')
window.configure(background = 'light blue')
window.iconbitmap('Downloads/student_4950.ico')

#Create a canvas
canvas= Canvas(window, width= 110, height= 100)
canvas.pack()

#Load an image in the script
img= (Image.open("Downloads/images.png"))

#Resize the Image using resize method
resized_image= img.resize((100,100), Image.ANTIALIAS)
new_image= ImageTk.PhotoImage(resized_image)

#Add image to the Canvas Items
canvas.create_image(10,10, anchor=NW, image=new_image)



####### Create a Table ########

#db=mysql.connector.connect(host="localhost",user="root",password="amruta95",database="python1")
#c=db.cursor()
#c.execute('create table person (Roll_No int , Name varchar(200),Class varchar(150), Marks float)')
#db.commit()
#db.close()

###### Insert Values (submit window)#######

def action():
    roll_no=e1.get()
    name=e2.get()
    clas=e3.get()
    marks=e4.get()
    
    db=mysql.connector.connect(host="localhost",user="root",password="amruta95",database="python1")
    c=db.cursor()
    c.execute("insert into person values('"+roll_no+"','"+name+"','"+clas+"','"+marks+"')")
    messagebox.showinfo("Information","Data Saved..!!" )
    db.commit()
    db.close()

    
########### Fetch Values (Results window) ######## 

def next():
    nw = Tk()
    nw.geometry("500x650")
    nw.minsize(500,650)
    nw.maxsize(500,650)
    nw.configure(background = 'light yellow')

    nw.title("Ranker's records")
    
    label=Label(nw,text='RESULT',font=('Helvetica',32,'bold'),fg='white',bg='red',padx=210,pady=15)
    label.grid(row=0,column=0 , columnspan=20)
    
    p1=Label(nw,text='Rank',font=('Helvetica',28,'bold'))
    p1.grid(row=1,column=2,padx=10,pady=10)
    
    p2=Label(nw,text='Student Name',font=('Helvetica',28,'bold'))
    p2.grid(row=1,column=0,padx=10,pady=10)
    
    
    db=mysql.connector.connect(host="localhost",user="root",password="amruta95",database="python1")
    c=db.cursor()
    c.execute("select * from person order by Marks desc limit 3 ")
    r=c.fetchall()
    for x in r:
        print(r)
    
    num=2
    for i in r:

        name=Label(nw,text=i[1],font=('Helvetica',20,'bold'),fg='black')
        name.grid(row=num,column=0,padx=10,pady=10)
        
        rank=Label(nw,text=i[2],font=('Helvetica',20,'bold'),fg='black')
        rank.grid(row=num,column=2,padx=10,pady=10)
        
        l1=Label(nw,text='First Rank',font=('Helvetica',20,'bold'),fg='black')
        l1.grid(row=2,column=2,padx=10,pady=10)
        
        l2=Label(nw,text='Second Rank',font=('Helvetica',20,'bold'),fg='black')
        l2.grid(row=3,column=2,padx=10,pady=10)
        
        l3=Label(nw,text='Third Rank',font=('Helvetica',20,'bold'),fg='black')
        l3.grid(row=4,column=2,padx=10,pady=10)
        
        
        num=num+1    
    db.commit()
    db.close()

########### Main Window ######

l1= Label(window,text='Student Management System',font=('Helvetica',20,'bold'))
l1.place(x=75,y=120)

l2= Label(window,text='Roll_No :=',font=('Helvetica',20,'bold'))
l2.place(x=30,y=220)

e1=Entry(window,width=30,bd=3)
e1.place(x=260,y=220)

l3= Label(window,text='Name :=',font=('Helvetica',20,'bold'))
l3.place(x=35,y=280)

e2=Entry(window,width=30,bd=3)
e2.place(x=260,y=280)

l4= Label(window,text='Class :=',font=('Helvetica',20,'bold'))
l4.place(x=30,y=340)

e3=Entry(window,width=30,bd=3)
e3.place(x=260,y=340)

l5= Label(window,text='Marks :=',font=('Helvetica',20,'bold'))
l5.place(x=30,y=400)

e4=Entry(window,width=30,bd=3)
e4.place(x=260,y=400)

button=Button(window,text='Results',fg='black',bg='yellow',font=('Helvetica',15,'bold'),width=15,command=next)
button.place(x=32,y=500)

button2=Button(window,text='Submit',fg='white',bg='blue',font=('Helvetica',15,'bold'),width=15,command=action)
button2.place(x=260,y=500)

window.mainloop()
```
